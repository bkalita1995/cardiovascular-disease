```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df=pd.read_csv('cardio_train.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['column']=df['id;age;gender;height;weight;ap_hi;ap_lo;cholesterol;gluc;smoke;alco;active;cardio']
```

```python
df.head()
```

```python
df['id']=df['column'].apply(lambda x: x.split(';')[0])
```

```python
df.head()
```

```python
df['age']=df['column'].apply(lambda x: x.split(';')[1])
df['gender']=df['column'].apply(lambda x: x.split(';')[2])
df['height']=df['column'].apply(lambda x: x.split(';')[3])
df['weight']=df['column'].apply(lambda x: x.split(';')[4])
df['ap_hi']=df['column'].apply(lambda x: x.split(';')[5])
df['ap_lo']=df['column'].apply(lambda x: x.split(';')[6])
df['cholestrol']=df['column'].apply(lambda x: x.split(';')[7])
df['gluc']=df['column'].apply(lambda x: x.split(';')[8])
df['smoke']=df['column'].apply(lambda x: x.split(';')[9])
df['alco']=df['column'].apply(lambda x: x.split(';')[10])
df['active']=df['column'].apply(lambda x: x.split(';')[11])
df['cardio']=df['column'].apply(lambda x: x.split(';')[12])
```

```python
df.head()
```

```python
df1=df.drop(['id;age;gender;height;weight;ap_hi;ap_lo;cholesterol;gluc;smoke;alco;active;cardio','column','id'],axis=1)
df1.head()
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
df1.isnull().sum()
```

```python
df1['age']=df1['age'].astype(int)
df1['height']=df1['height'].astype(float)
df1['weight']=df1['weight'].astype(float)
df1['ap_hi']=df1['ap_hi'].astype(int)
df1['ap_lo']=df1['ap_lo'].astype(int)
df1['cholestrol']=df1['cholestrol'].astype(int)
df1['gluc']=df1['gluc'].astype(int)
df1['smoke']=df1['smoke'].astype(int)
df1['alco']=df1['alco'].astype(int)
df1['active']=df1['active'].astype(int)
df1['cardio']=df1['cardio'].astype(int)
```

```python
df1.dtypes
```

```python
df1.head()
```

```python
df2=df1.copy()
```

```python
df2['age']=df2['age']/(12*30)
```

```python
df2.head()
```

```python
df2['age']=np.ceil(df2['age'])
df2.age.head()
```

```python
df2.describe()
```

```python
plt.figure(figsize=(20,25), facecolor='white' )
plotnumber=1

for column in df2:
    if plotnumber<=12:
        ax=plt.subplot(4,4,plotnumber)
        sns.distplot(df2[column])
        plt.xlabel(column,fontsize=20)
        
    plotnumber+=1
plt.show()
```

```python
bool_var=[var for var in df2 if df[var].nunique()==2]
df2[bool_var].head()
```

```python
num_var=[var for var in df2 if df2[var].dtypes !='O' and  var not in bool_var]
df2[num_var].head()
```

```python
plt.figure(figsize=(20,25), facecolor='white' )
plotnumber=1

for column in df2[num_var]:
    if plotnumber<=12:
        ax=plt.subplot(4,4,plotnumber)
        sns.boxplot(df2[num_var][column])
        plt.xlabel(column,fontsize=20)
        
    plotnumber+=1
plt.show()
```

```python
X = df2.drop(columns = ['cardio'])
y = df2['cardio']
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in X:
    if plotnumber<=9 :
        ax = plt.subplot(4,4,plotnumber)
        sns.stripplot(y,X[column])
    plotnumber+=1
plt.tight_layout()
plt.show()
```
